# 3.FEATURE BALANCING
## OVER SAMPLING

## RandomOverSampler

## Smote

```python
from imblearn.over_sampling import RandomOverSampler, SMOTE
```

```python
x = df.drop('EmploymentStatus', axis = 1)  
#entire dataset expect income column bcz we deleted and this updated data is stored in x
```

```python
  y = df.loc[: , ['EmploymentStatus']]  
#income column with all the rows stored in y
```

```python
### not majority - resample all the classes but the majority classes(expect majority class)
```

```python
y.value_counts()
```

```python
obj_ros = RandomOverSampler(sampling_strategy = 'not majority')
y_int = y.astype(int)
x_ros, y_ros = obj_ros.fit_resample(x,y_int)
```

```python
y_ros.value_counts() #After resampling
```

```python
## SMOTE
```

```python
### 50 % of miniority class is added here
```

```python
obj_smote = SMOTE(sampling_strategy = .5)

x_smote, y_smote = obj_smote.fit_resample(x,y_int)
```

```python
y_smote.value_counts()
```

# UnderSampling
## RandomUnderSampler
## EditedNearestNeighbours

```python
from imblearn.under_sampling import RandomUnderSampler
```

```python
obj_rus = RandomUnderSampler(sampling_strategy = 'not minority')
x_rus, y_rus = obj_rus.fit_resample(x,y_int)
```

```python
y_rus.value_counts()
```

```python
#EditedNearestNeighbours
from imblearn.under_sampling import EditedNearestNeighbours
```

```python
obj_enn = EditedNearestNeighbours(n_neighbors = 5)
x_enn, y_enn = obj_enn.fit_resample(x,y_int)
```

```python
y_enn.value_counts()
```
